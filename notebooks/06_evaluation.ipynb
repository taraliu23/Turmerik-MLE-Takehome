```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/final/evaluation.csv")

# df.head()
df.columns
```

```python

plt.figure(figsize=(12, 6))
plt.plot(df["Question ID"], df["Overall Score"], label="Overall Score (Before RAG)", marker='o', linestyle='--')
plt.plot(df["Question ID"], df["RAG Overall Score"], label="Overall Score (After RAG)", marker='o', linestyle='-')
plt.xlabel("Question ID")
plt.ylabel("Overall Score")
plt.title("Comparison of Overall Scores Before and After RAG")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("../data/final/overall_score_comparison.png", dpi=500)
```

```python

# compare the completenes score before and after RAG
plt.figure(figsize=(12, 6))
plt.plot(df["Question ID"], df["Completeness"], label="Completeness Score (Before RAG)", marker='o', linestyle='--')
plt.plot(df["Question ID"], df["Completeness after RAG"],
         label="Completeness Score (After RAG)", marker='o', linestyle='-')

plt.xlabel("Question ID")
plt.ylabel("Completeness Score")
plt.title("Comparison of Completeness Scores Before and After RAG")

plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("../data/final/completeness_score_comparison.png", dpi=500)
```

```python


df["Completeness Improvement"] = df["Completeness after RAG"] - df["Completeness"]
df["Overall Score Improvement"] = df["RAG Overall Score"] - df["Overall Score"]

total_completeness_improvement = df["Completeness Improvement"].sum()
total_completeness_improvement_percentage = (total_completeness_improvement / df["Completeness"].sum()) * 100

total_overall_score_improvement = df["Overall Score Improvement"].sum()
total_overall_score_improvement_percentage = (total_overall_score_improvement / df["Overall Score"].sum()) * 100

print(f"Total Completeness Improvement: {total_completeness_improvement}")
print(f"Total Completeness Improvement Percentage: {total_completeness_improvement_percentage:.2f}%")
print(f"Total Overall Score Improvement: {total_overall_score_improvement}")
print(f"Total Overall Score Improvement Percentage: {total_overall_score_improvement_percentage:.2f}%")

```
